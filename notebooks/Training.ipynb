```python
import pickle
import matplotlib.pyplot as plt
import os

# Directory where the checkpoints and losses are saved
CHECKPOINT_DIR = './checkpoints4'  # Update this if your checkpoint directory is different

# Path to the losses.pkl file
LOSSES_PATH = os.path.join(CHECKPOINT_DIR, 'losses.pkl')

# Check if the losses.pkl file exists
if not os.path.isfile(LOSSES_PATH):
    raise FileNotFoundError(f"Losses file not found at {LOSSES_PATH}. Please ensure the path is correct.")

# Load the losses from the pickle file
with open(LOSSES_PATH, 'rb') as f:
    losses_dict = pickle.load(f)

# Extract training and validation losses
train_losses = losses_dict['train_losses']
val_losses = losses_dict['val_losses']

# Create a range of epochs
epochs = range(1, len(train_losses) + 1)

# Plot the training and validation losses
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_losses, 'b-', label='Training Loss')
plt.plot(epochs, val_losses, 'r-', label='Validation Loss')
plt.title('Training and Validation Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

# Display the plot
plt.show()

# Optionally, save the plot to a file
PLOT_PATH = os.path.join(CHECKPOINT_DIR, 'loss_curve.png')
plt.savefig(PLOT_PATH)
print(f"Loss curve saved to {PLOT_PATH}")
```
